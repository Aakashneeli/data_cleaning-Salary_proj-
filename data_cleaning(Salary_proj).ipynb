```python
import pandas as pd
df = pd.read_csv("D:/kaggle/data_cleaned_2021.csv") 
df
```

```python
columns_to_del = ['Hourly', 'Employer provided', 'Lower Salary', 'Upper Salary', 'Avg Salary(K)', 'company_txt', 'Job Location', 'Age', 'Python', 'spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch', 'scikit', 'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo', 'google_an', 'job_title_sim', 'seniority_by_title', 'Degree']
df = df.drop(columns_to_del, axis = 1)
df
```

```python
#to do:
#salary parsing 
#company name text only 
#state field 
# age of company 
#parsing of jd(python , sql , etc)
```

```python
# SALARY PARSING

#removing -1 values 
df = df[df['Salary Estimate'] != '-1']
df
```

```python
#removing Glassdoor estimate , $ and k from salary estimate(using lambda function)
salary = df['Salary Estimate'].apply(lambda x : x.split('(')[0])
salary.head(210)
minus_k_dollars = salary.apply(lambda x: x.replace('K', '').replace('$', ''))
minus_k_dollars
```

```python
#making  a separate column for 'employer provided'  and 'per hour'
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer provided'] = df['Salary Estimate'].apply(lambda x : 1 if 'employer provided salary' in x.lower() else 0)
```

```python
#removing per hour and employer provided from minus_k_dollars
minus_perhr_emp = minus_k_dollars.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary', ''))
minus_perhr_emp
```

```python
# making columns fro min and max salary and average 
df['min_salary'] = minus_perhr_emp.apply(lambda x: int(x.split('-')[0].strip(':')))

df['max_salary'] = minus_perhr_emp.apply(lambda x: int(x.split('-')[1].strip(':')))

df['avg_salary'] = df['min_salary'] + df['max_salary']/2
df
```

```python
# COMPANY NAME TEXT ONLY
df['company_text_only'] = df.apply(lambda x : x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3], axis = 1)
df
```

```python
#STATES in which jobs are placed
df['job_state'] = df['Location'].apply(lambda x : x.split(',')[1])
df
#to see how many jobs are in a state 
df['job_state'].value_counts()
# creating a column to see if some jobs have same location and headquarters(might be an important data point in creation of algorithm)
df['same_state(Hq & Loc)'] = df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0, axis = 1)
df
```

```python
#age of company
df['age_of_company'] = df['Founded'].apply(lambda x: x if x<1 else 2023-x)
df
```

```python
#PARSING OF JD
#python
df['python_skill(y_or_n)'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['python_skill(y_or_n)'].value_counts()
#sql
df['sql_skill(y_or_n)'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df['sql_skill(y_or_n)'].value_counts()
#r studio
df['rstudio_skill(y_or_n)'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df['rstudio_skill(y_or_n)'].value_counts()
#spark
df['spark_skill(y_or_n)'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['spark_skill(y_or_n)'].value_counts()
#aws
df['aws_skill(y_or_n)'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['aws_skill(y_or_n)'].value_counts()
#excel
df['excel_skill(y_or_n)'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['excel_skill(y_or_n)'].value_counts()
```

```python

#till now we did data cleaning and some feature engineering
```

```python
#functions for making it easy (simplyfying the job roles)
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr.' in title.lower() or 'lead' in title.lower() or 'principle' in title.lower():
        return 'sr'
    elif 'jr' in title.lower() or 'junior' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['job_simplified'] = df['Job Title'].apply(title_simplifier)
df.job_simplified.value_counts()
```

```python
df['job_seniority'] = df['Job Title'].apply(seniority)
df.job_seniority.value_counts()
```

```python
#fix los angeles state
df['job_state'] = df['job_state'].apply(lambda x : x.strip() if x.strip().lower() != 'los angeles' else 'CA')
df.job_state.value_counts()
```

```python
#job description length
df['job_descp_len'] = df['Job Description'].apply(lambda x : len(x))
df.job_descp_len
```

```python
#competitors count 
df['num_competitors'] = df['Competitors'].apply(lambda x : len(x.split(',')) if x != -1 else 0)
```

```python
#hourly wage to annual
df['min_salary'] = df.apply(lambda x : x['min_salary']*2 if x['hourly'] == 1 else x['min_salary'], axis = 1)
```

```python
df['max_salary'] = df.apply(lambda x : x['max_salary']*2 if x['hourly'] == 1 else x['max_salary'], axis = 1)
df[df.hourly== 1][['min_salary', 'max_salary']]
```

```python
#removing of new line(\n) from company text
df['company_text_only'] = df['company_text_only'].apply(lambda x: x.replace('\n', ''))
df['company_text_only']
```

```python
#converting it into csv
df.to_csv('DS_salary_data_cleaned.csv', index = False)
pd.read_csv('DS_salary_data_cleaned.csv')
```
